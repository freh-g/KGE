# PARSE RESULTS OF SINGLE DISEASE CLASSES

```python
with open('../outputs/RandomVsEmbs/results_crossval_Metapath2Vec_final_SVC(C=10, gamma=0.1, probability=True).pickle','rb') as f:
    rm2v = pickle.load(f)
with open('../outputs/RandomVsEmbs/results_crossval_Random_SVC(C=10, gamma=0.1, probability=True).pickle','rb') as f:
    rr = pickle.load(f)
```

```python
models = ['Metapath2Vec','Random']
nofass =  [d[1]for d in rm2v.keys()]
data = [rm2v,rr]
metrics = list(list(rr.values())[0].keys())[2:]
index=[d[0] for d in rm2v.keys()]
multicol = pd.MultiIndex.from_product([models,columns],names=['model','metrics'])
df = pd.DataFrame(index=index,columns=multicol)
```

```python
for i,disease in enumerate(df.index):
    for model in models:
        if model =='Metapath2Vec':
            
            d = data[0]
            
        else:
            d = data[1]
        diction = list(d.items())[i][1]
        list_of_means_metrics = [np.mean(diction[m]) for m in metrics]
        df.loc[disease,model] =  list_of_means_metrics
        
    
        
    
```

```python
df
```

```python
df.insert(0,('','number of associations'),nofass)
```

```python
df.to_excel('../outputs/RandomVsEmbs/table2.xlsx')
```
